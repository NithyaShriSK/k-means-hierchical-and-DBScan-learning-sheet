```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
X, _ = make_blobs(n_samples=200, centers=4, random_state=42)
plt.scatter(X[:,0], X[:,1], s=30)
plt.title("Sample Data")
plt.show()
```

```python

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y_kmeans, cmap="tab10")
plt.title("KMeans Clustering")
plt.show()
```

```python
hier = AgglomerativeClustering(n_clusters=4)
y_hier = hier.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y_hier, cmap="tab10")
plt.title("Hierarchical Clustering")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y_dbscan, cmap="tab10")
plt.title("DBSCAN Clustering")
plt.show()
```
